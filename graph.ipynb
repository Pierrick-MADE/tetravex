```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stat = np.genfromtxt("log_s6.txt", delimiter=';')
```

```python
stat.shape
```

```python
iterations = np.arange(stat.shape[0])

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('iteration')
ax1.set_ylabel('score', color=color)
ax1.plot(iterations, stat[:,1], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Heat', color=color)  # we already handled the x-label with ax1
ax2.plot(iterations, stat[:,0], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
